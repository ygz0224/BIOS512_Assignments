Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width = 8, repr.plot.height = 8)

p2 <- ggplot( penguins, aes( x= body_mass_g ) )+
geom_histogram( binwidth=100, color="steelblue4", size=1.2, fill="skyblue3", alpha=0.7, na.rm=T )+
facet_wrap( ~year, nc=1, strip.position="top" )+
scale_x_continuous( name="Body mass in grams", limits = c(2500, 6500), breaks=seq(2500, 6500, 1000), labels=scales::comma )+
scale_y_continuous( name="Number of penguins", limits = c(0, 10), breaks=seq(0, 7.5, 2.5) )+
labs( title = 'Histograms of penguin body mass for measurements taken from 2007-2009', size=25 )+
theme_minimal()+
theme( plot.title = element_text( size=17 ), 
axis.title.x = element_text( size=20 ), 
axis.title.y = element_text( size=20 ), 
axis.text.x = element_text( size=15, color="grey40" ), 
axis.text.y = element_text( size=15, color="grey40" ), 
strip.text.x = element_text( size=15, color="grey40", face="bold", hjust=0.95 ) )

p2
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
p3 <- ggplot( penguins, aes( x= body_mass_g, y= sex ) )+
geom_point( position = position_jitter( w=0, h=.2 ), size=3, color=1, alpha=0.4, na.rm=T )+
facet_wrap( ~species, nc=1, strip.position="top" )+
scale_x_continuous( name="Body mass in grams", limits = c(2500, 6500), breaks=seq(3500, 5500, 1000), labels=scales::comma )+
scale_y_discrete( name="" )+
theme_minimal()+
theme( plot.title = element_text( size=17 ), 
axis.title.x = element_text( size=20 ), 
axis.title.y = element_text( size=20 ), 
axis.text.x = element_text( size=15, color="grey40" ), 
axis.text.y = element_text( size=15, color="grey40" ), 
strip.text.x = element_text( size=15, color="grey40", face="bold", hjust=0 ) )

p3
```
